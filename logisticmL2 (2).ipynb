```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
df1["Embarked"]=label_encoder.fit_transform(df1["Embarked"])
```

```python
df1["Sex"]=label_encoder.fit_transform(df1["Sex"])
```

```python
X = df1.drop(columns= 'Survived', axis= 1) # Storing features in X var except colum survived [independent variable]
Y = df1['Survived'] # Storing target var in Y var [dependent variable]
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
len(x_train),len(x_test)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train.values,y_train.values)
```

```python
x_train_prediction=model.predict(x_train.values)
training_data_accuracy=(x_train_prediction,y_train)
```

```python
training_data_accuracy
```

```python
x_test_prediction = model.predict(x_test.values)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```
